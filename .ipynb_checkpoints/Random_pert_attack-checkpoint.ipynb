```python
from __future__ import print_function
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torchvision import datasets, transforms
import numpy as np
import os
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt

# NOTE: This is a hack to get around "User-agent" limitations when downloading MNIST datasets
#       see, https://github.com/pytorch/vision/issues/3497 for more information
from six.moves import urllib
opener = urllib.request.build_opener()
opener.addheaders = [('User-agent', 'Mozilla/5.0')]
urllib.request.install_opener(opener)
```

```python
#pretrained_model = "data/lenet_mnist_model.pth"
use_cuda=False
```

```python
# LeNet Model definition
class LeNet(torch.nn.Module):
    def __init__(self, activation=None, num_classes=10, in_channels=1):
        super(LeNet, self).__init__()
        pool = torch.nn.AvgPool2d(2, 2, divisor_override=1)

        self.C1 = torch.nn.Conv2d(in_channels, 6, 5, padding=2)
        self.S2 = torch.nn.Sequential(pool, torch.nn.Conv2d(6, 6, 1, groups=6))
        self.C3 = torch.nn.Conv2d(6, 16, 5, groups=2)
        self.S4 = torch.nn.Sequential(pool, torch.nn.Conv2d(16, 16, 1, groups=16))
        
        self.C5 = torch.nn.Linear(16*5*5, 120)
        self.F6 = torch.nn.Linear(120, 84)
        self.out = torch.nn.Linear(84, num_classes)
        
        self.activation = torch.nn.Identity() if activation is None else activation
        self.num_classes = num_classes

    def forward(self, x):
        # (B, 1, 28, 28)
        x = self.activation(self.C1(x))
        # (B, 6, 28, 28)
        x = self.activation(self.S2(x))
        # (B, 6, 14, 14)
        x = self.activation(self.C3(x))
        # (B, 16, 10, 10)
        x = self.activation(self.S4(x))
        # (B, 16, 5, 5)
        x = x.reshape(-1, 16*5*5)
        # (B, 400)
        x = self.activation(self.C5(x))
        # (B, 120)
        x = self.activation(self.F6(x))
        # (B, 84)
        x = self.out(x)
        # (B, 10)
        return x

# Choose the activation function from the paper
activation = lambda x: 1.7159*torch.tanh(0.6667*x)
network = LeNet(activation=activation, in_channels=1)
print('Network parameters:', sum(p.numel() for p in network.parameters()))


# Define what device we are using
print("CUDA Available: ",torch.cuda.is_available())
device = torch.device("cuda" if (use_cuda and torch.cuda.is_available()) else "cpu")

# Initialize the network
loss_fn = nn.CrossEntropyLoss()
optimiser = optim.SGD(network.parameters(), lr=0.004)
file_name_basic_training = 'training_basic'
```

```python
# MNIST Test dataset and dataloader declaration
test_loader = torch.utils.data.DataLoader(
    datasets.MNIST('data', train=False, download=False, transform=transforms.Compose([
            transforms.ToTensor(),
            ])),
        batch_size=1, shuffle=True)

train_loader = torch.utils.data.DataLoader(datasets.MNIST('data', train=True,
                       transform=transforms.Compose([transforms.ToTensor(),]),
                       download=False),batch_size=64,shuffle=True,num_workers=4)
```

## Standard Training

```python
def plot_history(history):
    # Plot training and validation curves for avg. loss and accuracy
    fig, axes = plt.subplots(1, 2, figsize=(12, 4.5))
    data = [('Avg. Loss', history['train_loss'], history['val_loss']), 
            ('Accuracy', history['train_acc'], history['val_acc'])]
    for ax, (name, train_hist, val_hist) in zip(axes, data):
        ax.set_title(name)
        ax.set_xlabel('Epoch')
        ax.xaxis.set_major_locator(ticker.MaxNLocator(integer=True))
        ax.grid(linestyle='dashed')
        ax.plot(range(1, len(train_hist)+1), train_hist, label='Training')
        ax.plot(range(1, len(val_hist)+1), val_hist, label='Validation')
        ax.legend()
    
    plt.tight_layout()
    plt.show()
    plt.close(fig)

def train_val_basic(epoch,history):
    
    print('\n[ Train epoch: %d ]' % epoch)
    network.train()
    loss_train = 0
    correct_train = 0
    total = 0
    
    for idx_train, (x, y,) in enumerate(train_loader):
        optimiser.zero_grad()
        y_pred = network(x)
        loss = loss_fn(y_pred, y)
        loss.backward()
        optimiser.step()
        res = torch.argmax(y_pred, dim=-1)
        correct_train += torch.sum(res == y).data.cpu().numpy()
        
        total += x.shape[0]
        loss_train += loss.data.cpu().numpy() * x.shape[0]
        
    history['train_loss'].append(loss_train / total)
    history['train_acc'].append(correct_train / total)   
    print('\nTotal train accuarcy:', 100. * correct_train / total)
    print('Total train loss:', 100 * loss_train/total)
        
    loss_val = 0
    correct_val = 0
    total = 0
    print('\n[ Val epoch: %d ]' % epoch)
    network.eval()
    # Loop over all examples in test set
    for idx, (data, target) in enumerate(test_loader):

        # Forward pass the data through the model
        res = network(data)
        loss = loss_fn(res, target)
        res = torch.argmax(res, dim=-1)
        correct_val += torch.sum(res == target).data.cpu().numpy()
        total += data.shape[0]
        loss_val += loss.data.cpu().numpy() * data.shape[0]
    
    
    history['val_loss'].append(loss_val / total)
    acc = correct_val / total
    history['val_acc'].append(acc)
    print('\nTotal val accuarcy:', 100. * correct_val / total)
    print('Total val loss:',  100. * loss_val / total)
    
    if not os.path.isdir('checkpoint'):
        os.mkdir('checkpoint')
    
    torch.save(dict(epoch=epoch, model=network.state_dict(),\
                    opt=optimiser.state_dict(),history=history),'./checkpoint/' + file_name_basic_training)
    
    print('Model Saved!')
    return history
```

```python
history_train_basic = dict(train_loss=[], train_acc=[], val_loss=[], val_acc=[])
for epoch in range(0, 10):
    history_train_basic = train_val_basic(epoch, history_train_basic)
```

```python
plot_history(history_train_basic)
```

```python
def test_perturbations(epoch, epsilon, alpha ):
    # Accuracy counter
    adv_loss = 0
    adv_correct = 0
    total = 0
    adv_examples = []
    print('\n[ Test epoch: %d ]' % epoch)
    network.eval()
    # Loop over all examples in test set
    for idx, (data, target) in enumerate(test_loader):
        
        init_res = network(data)
        loss = loss_fn(init_res, target)
        init_res = torch.argmax(init_res, dim=-1)
        benign_correct += torch.sum(init_res == target).data.cpu().numpy()
        total += data.shape[0]
        benign_loss += loss.data.cpu().numpy() * data.shape[0]
        
        # If the initial prediction is wrong, dont bother attacking, just move on
        if init_res.item() != target.item():
            continue
        
        # Call Random Attack
        perturbed_data = perturb_new(epsilon, alpha, data, target)
        # Re-classify the perturbed image
        final_res = network(perturbed_data)
        loss = loss_fn(final_res, target)
        final_res = torch.argmax(final_res, dim=-1)
        adv_correct += torch.sum(final_res == target).data.cpu().numpy()
        total += perturbed_data.shape[0]
        adv_loss += loss.data.cpu().numpy() * perturbed_data.shape[0]
        
        if final_res.item() == target.item():
            adv_correct += 1
        else:
            # Save some adv examples for visualization later
            if len(adv_examples) < 5:
                adv_ex = perturbed_data.squeeze().detach().cpu().numpy()
                adv_examples.append( (init_res.item(), final_res.item(), adv_ex) )
    
    print(f"Epsilon: {epsilon}\tAlpha: {alpha}")
    
    final_acc = 100. * adv_correct / total
    final_loss =  100. * adv_loss / total
    
    # Return the accuracy and an adversarial example
    return final_acc, final_loss, adv_examples
```

## Perturbation function

```python
from torch.nn.functional import normalize
from torch.autograd import Variable
np.random.seed(42)
iters = 5
def perturb_new(epsilon, alpha_star,data,target):
    delta_ = normalize(torch.FloatTensor(np.random.uniform(0,epsilon,data.shape)))
    data_adv = data + delta_ 
    for i in range(iters):
        X_var = Variable(data_adv, requires_grad=True)
        y_var = Variable(torch.LongTensor(target))

        scores = network(X_var)
        loss = loss_fn(scores, y_var)
        loss.backward()

        grad = X_var.grad.data.cpu()
        
        # alpha star is multiplied to the infinity norm estimate(thus doubles as the step_size)
        data_adv = np.array(data_adv) + alpha_star * np.array(np.sign(grad)) 

        data_adv = np.clip(data_adv, data - (epsilon*alpha_star - 1), data + (epsilon*alpha_star - 1))
        data_adv = np.clip(data_adv, 0, 1) # ensure valid pixel range
    return data_adv
```

```python
def testing_samples(eps_alp):
    accuracies = []
    examples = []
    losses = []
    # Run test for each epsilon
    for eps, alp in eps_alp:
        acc,loss, ex = test_perturbations(1, eps, alp)
        print(f"Accuracy={acc}, Loss={loss}")
        accuracies.append(acc)
        examples.append(ex)
        losses.append(loss)
    return accuracies,losses,examples
```

```python
def sample_viewer(examples,eps_alp):
    cnt = 0
    plt.figure(figsize=(5*len(eps_alp)/2,len(eps_alp)*2))
    for i in range(len(eps_alp)):
        for j in range(len(examples[i])):
            ex = examples[i]
            cnt += 1
            plt.subplot(len(eps_alp),len(examples[0]),cnt)
            plt.xticks([], [])
            plt.yticks([], [])
            if j == 0:
                plt.ylabel(f"Eps: {eps_alp[i][0]},alpha: {eps_alp[i][1]}", fontsize=14)
            adv = ex[j][2]
            plt.title(f"{ex[j][0]}->{ex[j][1]}")
            plt.imshow(adv, cmap="gray")
    plt.tight_layout()
    plt.show()
```

## Adversarial Training 

```python
def train(model,epoch,epsilon, alpha_star,fname,history):
    
    print('\n[ Train epoch: %d ]' % epoch)
    model.train()
    train_loss = 0
    correct_train = 0
    total = 0
    for idx, (x, y,) in enumerate(train_loader):
        x_adv = perturb_new(epsilon, alpha_star,data=x,target=y) if (idx%10 == 0)  else x
        optimiser.zero_grad()
        y_pred = model(x_adv)
        loss = loss_fn(y_pred, y)
        loss.backward()
        optimiser.step()
        res = torch.argmax(y_pred, dim=-1)
        correct_train += torch.sum(res == y).data.cpu().numpy()
        
        total += x.shape[0]
        train_loss += loss.data.cpu().numpy() * x.shape[0]
    
    history['train_loss'].append(train_loss / total)
    history['train_acc'].append(correct_train / total)
    print('\nTotal train accuarcy:', 100. * correct_train / total)
    print('Total train loss:', 100 * train_loss/total)
    
    loss_val = 0
    correct_val = 0
    total = 0
    print('\n[ Val epoch: %d ]' % epoch)
    model.eval()
    # Loop over all examples in test set
    for idx, (data, target) in enumerate(test_loader):

        # Forward pass the data through the model
        adv_data = perturb_new(epsilon, alpha_star,data,target)
        res = model(adv_data)
        loss = loss_fn(res, target)
        res = torch.argmax(res, dim=-1)
        correct_val += torch.sum(res == target).data.cpu().numpy()
        total += data.shape[0]
        loss_val += loss.data.cpu().numpy() * data.shape[0]
    
    history['val_loss'].append(loss_val / total)
    acc = correct_val / total
    history['val_acc'].append(acc)
    print('\nTotal val accuarcy:', 100. * correct_val / total)
    print('Total val loss:',  100. * loss_val / total)
    
    if not os.path.isdir('checkpoint'):
        os.mkdir('checkpoint')
    
    torch.save(dict(epoch=epoch, model=model.state_dict(),\
                    opt=optimiser.state_dict(),history=history),'./checkpoint/' + fname)
    
    print('Model Saved!')
    return history
```

```python
epsilon=0.7
alpha_star=0.4
```

```python
adv_history = dict(train_loss=[], train_acc=[], val_loss=[], val_acc=[])
file_name = 'adversarial_training'
for epoch in range(0, 20):
    adv_history = train(network,epoch,epsilon, alpha_star,file_name,adv_history)
```
